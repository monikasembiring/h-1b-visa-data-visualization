## 4.2. Background Study: H-1B Petition Process

To understand the context more and be able to do a thorough analysis, we studied the general process of an H-1B visa petition. Every year, United States Citizenship and Immigration Services (USCIS) do a random selection process to select enough applicants to meet the H1B Cap of 65,000 for the Regular quota and 20,000 for the Master's quota (applicants with a master's degree or above). Not all petitions are subject to the cap. Based on [H1B Lottery Results: The Definitive Guide on 2022-23 Lottery Chances](https://www.immi-usa.com/h1b-lottery-2016-results-chances-process/), cap-exempt petitions are those that:
- Previously counted against the cap
- Are intended to extend or lengthen the H-1B worker's time in the country (visa extension)
- Change the employment terms for current H-1B employees (i.e., location, responsibilities, duties, etc.)
- Transfer an H-1B status to a different employers
- Allow the worker to be employed in a second position
- Are filed by a cap-exempt sponsor: an institute of higher education, a nonprofit organization associated with an institute of higher education, or a governmental research center.

Besides that, we also studied [H-1B Electronic Registration Process](https://www.uscis.gov/working-in-the-united-states/temporary-workers/h-1b-specialty-occupations-and-fashion-models/h-1b-electronic-registration-process) on the USCIS website. The process will generally consist of these steps and a timeline, as presented in Figure 2.
Timeline.png
1. Employers interested in sponsoring H-1B for individuals they intend to hire can begin creating H-1B registrant accounts at noon Eastern.
2. Employers will get at least 14 days to register the individuals with their company information electronically. They can edit, delete, and modify candidate details until registration closes.
3. USCIS will conduct computer-generated random selection, aka lottery. Then it will notify the selected registrants' employer/sponsors electronically, at last, to file H-1B petitions within 90 days. The unselected registrations are kept on reserve to make another selection as needed.
5. Starting from April 1st, employers must file H-1B petitions using Labor Condition Application (LCA) and all the information supporting the H-1B job offer for the selected registrant.
6. Then, USCIS will adjudicate the petition per their standard process and inform employers regarding their petition.

In the past, before Fiscal Year (FY) 2020, the random selection, aka H-1B lottery, was first conducted for the Master's quota. Then the unselected petitions from the Master's quota were put down in the General quota pool, and a random selection algorithm was run to fulfill the General quota cap. 

However, since FY 2020 season (March 2019), to give more priority to US masters students, USCIS changed the order of the lottery. These two flowcharts, Figure 3 and Figure 4, (adapted from [What is H1B Visa Lottery? FY 2023 Process? Hence, we will look at the wage for different levels of the jobs in the overall dataset to get an idea of how much certified and denied applicants are earning. 

There are 4 H1B Wage levels that are defined by the US Department of Labor based on the Occupational Employment Statistics(OES) program. 

> **H1B Wage Level 1 – Entry**: This wage level is for entry level H1B workers with basic understanding of job.


> **H1B Wage Level 2 – Qualified**: This wage level is for qualified workers with education, experience and who can perform moderate tasks with limited judgment.


> **H1B Wage Level 3 – Experienced** : This wage level is for experienced workers who have deeper experience that is attained with education and experience.


> **H1B Wage Level 4 – Fully Competent**: This wage level the most expert level of workers, who can do complex tasks and can exercise a lot of judgment in their activities.

Source: [Guide to H1B Prevailing Wages, Wage Levels – Find Minimum Wage?](https://redbus2us.com/h1b-minimum-wage-or-lca-prevailing-wage-for-a-position-in-an-area-in-usa/)
